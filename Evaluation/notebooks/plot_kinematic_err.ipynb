```python
p_range = np.concatenate((np.arange(0, 100, 10), np.arange(110, 170, 20)))
CNN_mean, CNN_iqr, HPS_mean, HPS_iqr = plotter.profile("eta")
CNN_noHPS_mean, CNN_noHPS_iqr = plotter_noHPS.profile("eta")
CNN_noPi0_mean, CNN_noPi0_iqr = plotter_noPi0.profile("eta")
```

```python
width = np.diff(p_range)
p_centre = p_range[:-1] + np.diff(p_range)/2

plt.figure(figsize=(6,6))

plt.errorbar(p_centre, HPS_mean, xerr=width/2, linestyle="", marker="o", label = "HPS")
plt.errorbar(p_centre, CNN_mean, xerr=width/2, linestyle="", marker="o", label = "CNN HPS DM 1")
plt.errorbar(p_centre, CNN_noHPS_mean, xerr=width/2, linestyle="", marker="o", label = "CNN HPS DM -1")
plt.errorbar(p_centre, CNN_noPi0_mean, xerr=width/2, linestyle="", marker="o", label = "CNN HPS DM 0")
plt.legend()
plt.ylabel("Mean Error")
plt.xlabel("Pi0 Momentum")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(6,6))

plt.errorbar(p_centre, HPS_iqr, xerr=width/2, linestyle="", marker="o", label = "HPS")
plt.errorbar(p_centre, CNN_iqr, xerr=width/2, linestyle="", marker="o", label = "CNN HPS DM 1")
plt.errorbar(p_centre, CNN_noHPS_iqr, xerr=width/2, linestyle="", marker="o", label = "CNN HPS DM -1")
plt.errorbar(p_centre, CNN_noPi0_iqr, xerr=width/2, linestyle="", marker="o", label = "CNN HPS DM 0")
plt.legend()
plt.ylabel("IQR")
plt.xlabel("Pi0 Momentum")
plt.grid()
plt.show()
```
